## 3) Defina una red neuronal vía *Keras*  
---

En keras podremos definir de manera bastante cómoda y flexible una red neuronal, la forma en que lo haremos será la más sencilla, de manera secuencial, una capa tras otra.


1. Defina un "container" que será usado para encapsular nuestro modelo
```python
from keras.models import Sequential
model = Sequential()
```

2. Agregue una capa de 10 neuronas, los datos tienen una dimensión de entrada $D=5$, además añadir función de activación *tanh*.
```python
from keras.layers import Dense, Activation
model.add(Dense(units=10, input_shape=(D,)))
model.add(Activation('tanh'))
```

3. Agregue una capa para realizar regresión unidimensional.
```python
model.add(Dense(units=1)) #default is linear
```

4. ¿Cuántos parámetros tiene la red? Visualice
```python
model.summary()
weights = model.get_weights()
```
---

## 4) Entrene el modelo!
1. Simule datos artificiales para comprobar que su modelo funciona.
```python
import numpy as np
N = 100
X = np.ones((N,D))
Y = np.zeros((N,1))
```

2. Compile el modelo ¿Qué significa éste paso?
```python
from keras.optimizer import SGD
opt = SGD() #default params
model.compile(loss='mse', optimizer=opt)
```

3. Entrene
```python
model.fit(X,Y)
```
