```python
from scipy import misc
from sklearn import cluster
import numpy as np
#face = misc.face(gray=True).astype(np.float32)
face = lena
X = face.reshape((-1, 1))  # We need an (n_sample, n_feature) array
K = k_means = cluster.KMeans(n_clusters=5)  # 5 clusters
k_means.fit(X) 
#KMeans(...)
values = k_means.cluster_centers_.squeeze()
labels = k_means.labels_
face_compressed = np.choose(labels, values)
face_compressed.shape = face.shape
```

```python
labels = labels.reshape(face.shape)
plt.figure(figsize=(5,5))
plt.imshow(face_compressed, cmap='gray')
for l in range(5):
    plt.contour(labels == l, contours=1, colors=[plt.cm.spectral(l / float(5)),])
plt.show()
```

```python
import numpy, sklearn, scipy
import matplotlib
from matplotlib import pyplot as plt
from sklearn.feature_extraction import image
from sklearn.cluster import spectral_clustering
def lena():
    import pickle, os
    fname = 'lena.dat'
    f = open(fname,'rb')
    lena = np.array(pickle.load(f))
    f.close()
    return lena
lena = lena()
print lena.shape
graph = image.img_to_graph(lena)
graph.data = numpy.exp(-5*graph.data / lena.std())
labels = spectral_clustering(graph, n_clusters=11, n_components=11,assign_labels="kmeans", eigen_solver="amg", n_init=1)

labels = labels.reshape(lena.shape)
plt.figure(figsize=(5,5))
plt.imshow(lena, cmap='gray')
for l in range(11):
    plt.contour(labels == l, contours=1, colors=[plt.cm.spectral(l / float(11)),])
plt.show()
```
