Para poder trabajar en el curso se necesitará instalar librerías para Python (o R en su defecto), por lo que se recomienda instalarlas a través de *anaconda* (para Windows y sistemas Unix) en un entorno virtual o usando una plataforma virtual como Colaboratory de Google, donde podrán elegir su versión de Python (o R). 

> [Jupyter Notebook](https://jupyter.org/) es una herramienta que ofrece poder juntar código con texto y otro tipo de estructuras, html, latex, imágenes, vídeos, audios, entre otros. Esto se permite a través de utilizar IPython un Python interactivo donde la ejecución del código es a través de celdas, olvidándonos de un código gigante secuencial. Jupyter Notebook nos crea un servidor local donde podremos acceder a este IPython de una manera mucho mas cómoda, utilizando el navegador en nuestro computador (usualmente accediendo a localhost). 

> Para Python, las librerías básicas de *computer science* a utilizar serán *numpy*, *matplotlib*, *pandas*, *scipy*, *seaborn*, además de claramente *jupyter*. Mientras que algunas más avanzadas para data science corresponden a __[*sklearn*](http://scikit-learn.org/stable/)__ (para tecnicas clásicas de machine learning) y __[*keras*](https://keras.io/)__ o __[*pytorch*](https://pytorch.org/)__ (para para redes neuronales).

## Opción 1: Entorno/ambiente virtual 
---

<img src="https://i.ytimg.com/vi/Q0jGAZAdZqM/maxresdefault.jpg" title="resnet" width="45%"/>

Descargar e instalar __[Anaconda](https://www.anaconda.com/distribution/)__.

> Crear el entorno virtual no es obligación, es algo recomendable para tener empaquetado diferentes versiones de librerías en un mismo sitio.

* Instalar entorno virtual de **Python**
```
conda create -n py_env python=version
```
donde "*version*", la versión de Python que desea utilizar. Si está en Windows, se necesita Python 3. Dentro de las opciones de Python 3 se recomienda 3.3-3.7 debido a las dependencias con las librerías a utilizar.
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/768px-Python-logo-notext.svg.png" width="15%"/>


* Instalar entorno virtual de **R**
```
conda create -n r_env r-essentials r-base
```
R está diseñado para trabajo puramente estadístico, a diferencia de Python que depende de librerías estadísticas. Observación: **Los índices en R empiezan con 1, no con 0**

## Instalaciones y factores útiles


### Para instalar librerías de *Python* en el ambiente de Anaconda
---

Acceder al ambiente creado
```
source activate py_env
```
ó
```
conda activate py_env
```

* Vía pip (+*Colaboratory*)
    ```
    python -m pip install jupyter numpy scipy seaborn ipykernel pandas matplotlib sklearn
    ```
    > Librerías avanzadas de data science:
    ```
    python -m pip install tensorflow-gpu==version keras
    ```
    
    Afortunadamente *Colaboratory* viene con muchas librerías compatibles ya instaladas, pero si desea instalar o cambiar de versión alguna puede ejecutar:    
    ```
    !pip install libreria==version
    ```   

* Vía conda
    ```
    conda install jupyter numpy scipy seaborn ipykernel pandas matplotlib sklearn 
    ```
    > Librerías avanzadas de data science:
    ```
    conda install tensorflow-gpu==version keras-gpu
    ```
    
* Abrir un Jupyter
    ```
    jupyter notebook
    ```
    
Para salir del entorno
```
source/conda deactivate
```

*No olvide revisar la compatibilidad de las librerías entre sí o  con Cuda (si es que va a utilizar GPU).*

### Para instalar librerías de *R* en el ambiente Anaconda

Acceder al ambiente creado
```
source activate r_env
```
ó
```
conda activate r_env
```
* Abrir un Jupyter
    ```
    jupyter notebook
    ```
* Instalar paquete dentro de Jupyter Notebook o dentro de la consola luego de ejecutar "*R*" en el ambiente
    ```
    install.packages('nombre_paquete')
    ```

### Utilizar GPU propia (sólo si requiere)
---

Para ésto deberá instalar __[Cuda](https://developer.nvidia.com/cuda-toolkit-archive)__ en su computador. Cuda es una librería en *C* para utilizar la GPU de manera paralela. La versión a instalar dependerá fuertemente de las librerías que utilizaremos y su versión. Debe estar atento a la versión de cuda instalado, ya que debe ser compatible con las versiones de librerías. Por ejemplo [tensorflow-gpu](https://www.tensorflow.org/install/source) o [tensorflow + keras](https://docs.floydhub.com/guides/environments/).


Para verificar que *Cuda* está instalado en su computador y tiene las dependencias, deberá ejecutar:
```
nvcc --version
```

Para verificar la información de la *GPU*, si es que su computador tiene, deberá ejecutar:
```
nvidia-smi
```

### Verificar si *tensorflow* reconoce GPU
---
```python
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```


### Instalar librería temporal en *Colaboratory*
---
A pesar que Colaboratory viene con muchas librerías instaladas, uno puede ejecutar comandos de consola con el sígno de exclamación, entonces para instalar vía *pip* sería algo como:
```python
!pip install opencv
```


### Utilizar datos en *Colaboratory*
---

1. Montar el *path* de su Google Drive
```python
from google.colab import drive
drive.mount('/content/drive')
```
2. Descargar datos a través de comandos en consola con jupyter
```python
!wget https://link.com
```
3. Suba los datos a través de la interfaz de Colaboratory
```python
from google.colab import files
uploaded = files.upload()
```

## Buenas prácticas y tips
---

Algunos *tips* útiles del framework Jupyter.

* **ctrl + shift + p** = abre buscador de comandos  
* **ctrl + shift + -** = divide la celda actual desde donde se encuentra el cursor  
* **Esc + f** = find and replace en código sin afectar salidas  
* **shift + M** = merge de celdas  

> **Otros atajos en __[Jupyter Notebook tips](https://www.dataquest.io/blog/jupyter-notebook-tips-tricks-shortcuts/)__**

En Jupyter se puede combinar texto con código, como por ejemplo un "Hola Mundo" en Python.

```python
print "Hola Mundo"
```

````python
print "Hola Mundo"
````

También definir variables (arreglos, listas, matrices) y visualizarlas a través de la consola interactiva de python (IPython)

````python
aux = [1,2,3,5,10]
aux
````

````python
#o en R
aux <- c(1,2, 3, 5, 10)
aux
````

Se pueden ejecutar comandos de consola con el signo de exclamación

````python
!cat /proc/cpuinfo
````

Si no se conoce la documentación de alguna función se puede añadir el signo de pregunta para consultar

|Header 1| Header 2|Header 3|
|---|---|---|
|Fila1|Fila1|Fila1|
|Fila2|Fila2|Fila2|

$$

$$

escribir en **negritas** y *cursiva*.
