## b) Preprocesar datos (secuencia)

```python
def create_dataset(secuencia,lag =1): 
    #preprocesamiento asumiendo dataset es una secuencia
    dataX = []
    dataY =[]
    for t_plus_1 in range(lag,len(secuencia)):
        #se crea el Y (target)
        predecir = secuencia[t_plus_1]
        dataY+= list(predecir) 
        
        #se crea el X (inputs) columnas para predecir Y
        columnasX = []
        for i in np.arange(lag,0,-1): #para los valores anteriores al t_plus_1 durante un lag
            columnasX += list( secuencia[t_plus_1-i] )
        dataX.append(columnasX)
    return np.array(dataX),np.array(dataY)
#create_dataset(stream_train_scaled)
```

En la celda anterior se crea una función que crea las secuencias que serán utilizadas para predecir los datos.

```python
lag = 3
trainX, trainY = create_dataset(df_train, lag)

length = 12
sec = df_train[:length]
print "Secuencia hasta los ",length
print sec.T[0]

#para visualizar que se hizo bien
datos = []
for m in range(length-lag):
    x= trainX[m]
    y= trainY[m]
    aux = ["-"] + [ "X"+str(i) for i in range(lag)] +["Y"]
    datos += [ ["Dato " + str(m)]+ list(x) + [y] ] 
    
from tabulate import tabulate
print tabulate(datos, headers = aux, tablefmt = "fancy_grid")
```

En la celda anterior se presenta el cómo será el formato que tendrán los datos para ser trabajados como *input* (características) y como *target* (predecir), con un lag de 3, lo cual es la cantidad de datos en la secuencia que serán utilizados para predecir un valor futuro.  
Se puede ver el ejemplo en que la secuencia de 12 meses es convertida a 9 ejemplos con un lag de 3, es decir, 3 datos de la secuencia son utilizados para predecir el siguiente valor en esta.

## h) Variar numero de bloques

```python
def computar_error(modelo,X,Y):
    Ypredict = modelo.predict(X)
    #invertir escalamiento
    Ypredict = scaler.inverse_transform(Ypredict)
    Yreal = scaler.inverse_transform([Y])

    return np.sqrt(mean_squared_error(Yreal[0], Ypredict[:,0]))
```

```python
def create_model(bloques,lag):
    # create your model using this function
    model = Sequential()
    model.add(LSTM(units=bloques, input_shape=(None,lag), activation='tanh', recurrent_activation='sigmoid'))
    model.add(Dense(1, activation = 'linear'))

    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

nb=range(4,13,2)

for bloques in nb:                                                                    
    kf = KFold(n_splits=5)
    error_cv = []
    tiempo_entrenarse = []
    for cv_train, cv_val in kf.split(trainX):#por cada fold
        model = create_model(bloques,lag=3)
       
        start_time = time.time()
        #entrenar    
        model.fit(trainX[cv_train], trainY[cv_train], epochs=100, batch_size=1, verbose=0)
        
        tiempo_entrenarse.append(time.time() - start_time )

        #computar error
        valScore = computar_error(model,trainX[cv_val],trainY[cv_val]) 
        error_cv.append(valScore)
    print('VAL Score con %d bloques: %.2f RMSE ---demorandose: %f seg' % 
          (bloques,np.mean(error_cv),np.mean(tiempo_entrenarse)))
```

En la celda anterior se varía la cantidad de bloques (unidades) de 4 a 12 de dos en dos, en la capa escondida con LSTM de la primera red definida, calculando el RMSE mediante 5-fold cross validation, mostrándo el tiempo que demora en entrenar cada red. El tiempo en entrenar la red no varía fuertemente respecto a la cantidad de bloques en esos rangos.  
Se puede ver una tendencia, exceptuando por 10 bloques, de que al aumentar los bloques el error disminuye. El menor *score* en cross validation es el obtenido por la red con 12 bloques, con un valor de 22,18.

### Se escoge el mejor modelo: 12 bloques

```python
#mejor modelo
model = Sequential()
model.add(LSTM(units=12, input_shape=(None,lag), activation='tanh', recurrent_activation='sigmoid'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=0)

#computar error
trainScore = computar_error(model,trainX,trainY)
testScore = computar_error(model,testX,testY)
print('Train Score: %.2f RMSE' % (trainScore))
print('Test Score con 12 bloques: %.2f RMSE' %(testScore) )
```

El mejor modelo con cross validation es probado con el conjunto de pruebas y presentado su error RMSE. Se puede ver que este error es menor que el alcanzado anteriormente con el modelo con 4 bloques, por lo que añadir bloques aumenta el nivel de cómputo de la red mejorando su resultado.

## i) Variar el lag

```python
def generar_secuencias(datos_entrenamiento, datos_pruebas,lag=1):
    trainX, trainY = create_dataset(datos_entrenamiento, lag)
    testX, testY = create_dataset(datos_pruebas, lag)
    #reshape
    trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
    testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
    return trainX,trainY, testX, testY
```

```python
lags = [1,2,3,4]

for lag in lags:
    trainX,trainY, testX,testY = generar_secuencias(stream_train_scaled,stream_test_scaled,lag = lag)

    kf = KFold(n_splits=5)
    error_cv = []
    tiempo_entrenarse = []
    for cv_train, cv_val in kf.split(trainX):#por cada fold
        #crear modelo
        model = Sequential()
        model.add(LSTM(4, input_shape= (None,lag) , activation='tanh', recurrent_activation='sigmoid'))
        model.add(Dense(1, activation = 'linear'))
        
        start_time = time.time()
        #entrenar    
        model.compile(loss='mean_squared_error', optimizer='adam')
        model.fit(trainX[cv_train], trainY[cv_train], epochs=100, batch_size=1, verbose=0)
        
        tiempo_entrenarse.append(time.time() - start_time )
        
        #computar error
        valScore = computar_error(model,trainX[cv_val],trainY[cv_val]) 
        error_cv.append(valScore)
    print('VAL Score con lag %d : %.2f RMSE ---demorandose: %f seg' % 
          (lag,np.mean(error_cv),np.mean(tiempo_entrenarse)))
```

En la celda anterior se varía el lag para generar las secuencias desde 1 hasta 4, pudiendo ver su error RMSE con cross validation de 5 fold, asociado con su tiempo de entrenar la red. Se puede ver que al aumentar el lag, exceptuando por lag 1, el tiempo de entrenar la red aumenta, debido a que se tienen más características con las cuales se debe trabajar la red.  
El mejor modelo escogido es con lag igual a 1, con RMSE 22,80, es decir, utiliza 1 característica para predecir, el valor anterior a un dato en la secuencia.

### Se escoge la mejor: lag igual a 1

```python
#elegir la mejor y ver...
lag = 1
trainX,trainY, testX,testY = generar_secuencias(stream_train_scaled,stream_test_scaled,lag = lag)
model = Sequential()
model.add(LSTM(4, input_shape= (None,lag) , activation='tanh', recurrent_activation='sigmoid'))
model.add(Dense(1, activation = 'linear'))

#entrenar    
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=0)

#computar error
trainScore = computar_error(model,trainX,trainY)
testScore = computar_error(model,testX,testY)
print('Train Score: 22.80 RMSE')
print('Test Score con lag 1: 53.73 RMSE')
```

En la celda anterior se presenta el mejor modelo sobre el conjunto de pruebas, alcanzando disminuir levemente el error RMSE a 53,73, por lo que para este problema se puede ver que utilizar unicamente un dato en la secuencia para predecir la siguiente obtiene un mejor resultado, siendo que al utilizar 3 como fue el primer caso presentado puede empeorar el conocer el siguiente elemento en la secuencia.

```python
#entrenamiento
trainPredict = model.predict(trainX)
#invertir escalamiento
trainPredict = scaler.inverse_transform(trainPredict)

#pruebas
testPredict = model.predict(testX)
#invertir escalamiento
testPredict = scaler.inverse_transform(testPredict)

# shift train predictions for plotting
trainPredictPlot = np.empty_like(dataframe.values)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[lag:len(trainPredict)+lag, :] = trainPredict

# shift test predictions for plotting
testPredictPlot = np.empty_like(dataframe.values)
testPredictPlot[:, :] = np.nan
testPredictPlot[(len(trainPredict)+2*lag):, :] = testPredict

plt.figure(figsize=(15,7))
plt.plot(dataframe.values,'b', label = "Dataset Real")
plt.plot(trainPredictPlot,'m', label= "Entrenamiento Predecido")
plt.plot(testPredictPlot,'r', label= "Pruebas Predecido")
plt.ylim(0)
plt.ylabel("Numero de vuelos")
plt.xlabel("Tiempo")
plt.title("Numero de vuelos a lo largo del tiempo")
plt.legend(loc="upper left")
plt.show()
```

En el gráfico anterior se puede comparar el valor real con el predecido por el último modelo entrenado (con lag igual a 1), obteniendo curvas que son bastante similares entre sí, por lo que se puede ver el gran poder de cómputo que tiene este tipo de red, donde unicamente al utilizar el valor anterior la secuencia, puede predecir el siguiente.

## j) Comparar desempeño

```python
def modelo_gru(inp):
    model = Sequential()
    model.add(GRU(units=4, input_shape=(None,inp), recurrent_initializer='orthogonal', activation='tanh'))
    model.add(Dense(1, activation = 'linear'))
    return model
def modelo_simple(inp):
    model = Sequential()
    model.add(SimpleRNN(units=4, input_shape=(None,inp), recurrent_initializer='orthogonal',activation='tanh'))
    model.add(Dense(1, activation = 'linear'))
    return model  

trainX,trainY, testX,testY = generar_secuencias(stream_train_scaled,stream_test_scaled,lag = 3)
    
#modelo gru
model = modelo_gru(trainX.shape[2])

start_time = time.time()
#entrenar    
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=0)

print('MODELO GRU')
print('Tiempo en entrenar: %f'%(time.time() - start_time ))
print('Train Score: %.2f RMSE' % (computar_error(model,trainX,trainY)))
print('Test Score: %.2f RMSE' % (computar_error(model,testX,testY)))

#modelo simpleRNN
model = modelo_simple(trainX.shape[2])

start_time = time.time()
#entrenar    
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=0)

print('MODELO SIMPLERNN')
print('Tiempo en entrenar: %f'%(time.time() - start_time ))
print('Train Score: %.2f RMSE' % (computar_error(model,trainX,trainY)))
print('Test Score: %.2f RMSE' % (computar_error(model,testX,testY)))
```

En las celdas anteriores se entrenan otros modelos de redes recurrentes (GRU y Simple RNN), pudiendo observar sus errores RMSE asociados sobre conjunto de entrenamiento y de pruebas. Se puede ver que el modelo GRU es el que obtiene un peor desempeño, superior al LSTM definido al comienzo de esta sección. Por otro lado el modelo Simple RNN botiene un desempeño mejor, disminuyendo el error que obtiene el modelo con LSTM, por lo que se puede ver que para este problema lo mejor es una red recurrente simple, con una sola conexión en el tiempo.

Se puede ver que el valor de error alcanzando sobre el conjunto de pruebas es menor al utilizar la representación con *time step* 1 y 3 características, debido a que esta última no aprovecha la estructura de la LSTM, pudiendo ser resuelta unicamente con una red *Feed Forward*, por lo que esta forma de representar los datos es mejor, ya que es la adecuada para trabajar la secuencia, teniendo un dato en el tiempo por *time step* con sus características. Se observa por otro lado este error sigue siendo mayor que la red recurrente simple, dato curioso que solo apunta a que puede ser por la falta de datos o que para este problema una red simple es lo adecuado para resolverlo debido a los pocos *time step* presentes en la arquitectura.

## k) memoria Batches

```python
def computar_error(modelo,X,Y,batch):
    Ypredict = modelo.predict(X,batch_size=batch)
    #invertir escalamiento
    Ypredict = scaler.inverse_transform(Ypredict)
    Yreal = scaler.inverse_transform([Y])

    return np.sqrt(mean_squared_error(Yreal[0], Ypredict[:,0]))
```

```python
batch_size = 1
model = Sequential()
model.add(LSTM(4, batch_input_shape=(batch_size, lag, 1), stateful=True,activation='tanh', recurrent_activation='sigmoid'))
model.add(Dense(1,activation='linear'))
model.compile(loss='mean_squared_error', optimizer='adam')

start_time= time.time()
for i in range(100):
    model.fit(trainX, trainY, epochs=1, batch_size=batch_size, verbose=0, shuffle=False)
    model.reset_states()
print('Tiempo en entrenar: %f'%(time.time() - start_time ))
print('Train Score: %.2f RMSE' % (computar_error(model,trainX,trainY,batch_size)))
print('Test Score: %.2f RMSE' % (computar_error(model,testX,testY,batch_size)))
```

En la celda anterior se entrena variando la red definida anteriormente, con *time step* de 3 y 1 característica, a una que ahora tiene memoria entre *batches*, esto puede verse en el parámetro *stateful* al definir el modelo anterior y que en cada *fit* de un *epoch* se deben resetear los estados. Lo que hace esto es guardar los pesos de la red para el próximo elemento del *batch* que tenga ese índice. Esto ayuda a que al utilizar *batchs* de tamaño 1 pueda "guardar" el estado y no volver a calcular todo denuevo, esto es enfocado a que los datos sean entregados secuencialmente, por lo que el método puede aprender las dependencias a través del tiempo en los datos.    
Esta forma de entrenar claramente entrega el mejor resultado comparado a las redes definidas anteriormente, incluso que la red recurrente simple (Simple RNN), debido a que esta sería la forma correcta de entrenar una red recurrente por todo lo teórico anteriormente dicho, de que almacenar el estado de cada ejemplo en la secuencia ayuda a encontrar las dependencias a través de este tiempo.

Al utilizar tamaño de *batch* igual a 3, en la misma red definida anteriormente, se obtiene un valor de error peor, sobre el conjunto de entrenamiento y de pruebas, por lo que puede ser que para este problema esta forma de entrenar mediante 3 elementos en un *batch* a la vez no sea la correcta, debido a que empeora la predicción. Otra causa puede ser ya que se cuenta con pocos datos, por lo que al utilizar un tamaño de *batch* mas grande empeorará la convergencia del algoritmo.
