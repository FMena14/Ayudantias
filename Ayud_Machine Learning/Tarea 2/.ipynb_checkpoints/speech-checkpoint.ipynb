# Google Audioset
link: https://research.google.com/audioset/

```python
import pandas as pd
import numpy as np
import tensorflow as tf


data_path = './google/audioset_v1_embeddings/bal_train'

session_conf = tf.ConfigProto(
    device_count={'GPU' : 0},
    allow_soft_placement=True,
    log_device_placement=False
)
```

```python

audio_record = data_path+'/_1.tfrecord'

#thanks to https://github.com/ibab/tensorflow-wavenet/issues/237
vid_ids = []
labels = []
audio_embedding = []
start_time_seconds = [] # in secondes
end_time_seconds = []
feat_audio = []
count = 0
for example in tf.python_io.tf_record_iterator(audio_record):
    tf_example = tf.train.Example.FromString(example)
    #print(tf_example)
    vid_ids.append(tf_example.features.feature['video_id'].bytes_list.value[0].decode(encoding='UTF-8'))
    labels.append(tf_example.features.feature['labels'].int64_list.value)
    start_time_seconds.append(tf_example.features.feature['start_time_seconds'].float_list.value[0])
    end_time_seconds.append(tf_example.features.feature['end_time_seconds'].float_list.value[0])

    tf_seq_example = tf.train.SequenceExample.FromString(example)
    n_frames = len(tf_seq_example.feature_lists.feature_list['audio_embedding'].feature)

    sess = tf.InteractiveSession()
    rgb_frame = []
    audio_frame = []
    # iterate through frames
    for i in range(n_frames):
        audio_frame.append(tf.cast(tf.decode_raw(
                tf_seq_example.feature_lists.feature_list['audio_embedding'].feature[i].bytes_list.value[0],tf.uint8)
                       ,tf.float32).eval())
    sess.close()
    feat_audio.append([])
    feat_audio[count].append(audio_frame)
    count+=1
```

## create dataframe


```python
from os import walk

files = []
for (dirpath, dirnames, filenames) in walk(data_path):
    files.extend(filenames)
    break
print files[:10]
print len(files)
```

```python
df = pd.DataFrame()#columns=["VideoId","Startsec","Endsec","Data","label"])
df.head()
```

```python
for record in files[:50]: #all
    audio_record = data_path+'/'+record

    #thanks to https://github.com/ibab/tensorflow-wavenet/issues/237
    vid_ids = []
    labels = []
    audio_embedding = []
    start_time_seconds = [] # in secondes
    end_time_seconds = []
    feat_audio = []
    count = 0
    for example in tf.python_io.tf_record_iterator(audio_record):
        tf_example = tf.train.Example.FromString(example)
        #print(tf_example)
        vid_ids.append(tf_example.features.feature['video_id'].bytes_list.value[0].decode(encoding='UTF-8'))
        labels.append(tf_example.features.feature['labels'].int64_list.value)
        start_time_seconds.append(tf_example.features.feature['start_time_seconds'].float_list.value[0])
        end_time_seconds.append(tf_example.features.feature['end_time_seconds'].float_list.value[0])

        tf_seq_example = tf.train.SequenceExample.FromString(example)
        n_frames = len(tf_seq_example.feature_lists.feature_list['audio_embedding'].feature)

        sess = tf.InteractiveSession()
        rgb_frame = []
        audio_frame = []
        # iterate through frames
        for i in range(n_frames):
            audio_frame.append(tf.cast(tf.decode_raw(
                    tf_seq_example.feature_lists.feature_list['audio_embedding'].feature[i].bytes_list.value[0],tf.uint8)
                           ,tf.float32).eval())
        sess.close()
        feat_audio.append(0)
        feat_audio[count] = audio_frame
        count+=1
        
        #print np.asarray(audio_frame).shape
        
    #a√±ade al final de la fila
    df = df.append(zip(*[vid_ids,start_time_seconds,end_time_seconds,np.asarray(feat_audio),labels]))
print "fin"
```

```python
df.columns = ["VideoId","Startsec","Endsec","Data","label"]
df.head()
```
